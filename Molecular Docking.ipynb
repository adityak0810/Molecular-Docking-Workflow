```python
import numpy as np

# Path to pocket12 PDB
pdb_path = "/content/fpocket/pocket12_atm.pdb"

# Parse coordinates
coords = []
with open(pdb_path, 'r') as f:
    for line in f:
        if line.startswith("ATOM") or line.startswith("HETATM"):
            x = float(line[30:38])
            y = float(line[38:46])
            z = float(line[46:54])
            coords.append([x, y, z])

coords = np.array(coords)
center = coords.mean(axis=0)

print(f"Center coordinates (x, y, z): {center}")
```

```python
import py3Dmol

# Corrected file paths
receptor_path = "/content/fpocket/2J1X_out/pockets/receptor12_h.pdbqt"
ligand_path = "/content/fpocket/2J1X_out/pockets/docked_ligand12.pdb"

# Initialize viewer
view = py3Dmol.view()

# Add receptor model
with open(receptor_path) as f:
    view.addModel(f.read(), "pdbqt")
view.setStyle({'model': 0}, {'cartoon': {'color': 'lightblue'}})

# Add ligand model
with open(ligand_path) as f:
    view.addModel(f.read(), "pdb")
view.setStyle({'model': 1}, {'stick': {'colorscheme': 'greenCarbon'}})

# Final render
view.zoomTo()
view.show()
```
